```python
# ════════════════════════════════════════════════════════════════════
# CARDIOVASCULAR DISEASE PREDICTION — FULL PIPELINE
# Train/Test Plot · SMOTE Plot · Mean±SD · Accuracy+AUPRC · McNemar
# ════════════════════════════════════════════════════════════════════

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from itertools import combinations
from statsmodels.stats.contingency_tables import mcnemar
from sklearn.model_selection import StratifiedKFold, train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, roc_auc_score, average_precision_score
)
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import BernoulliNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import (
    RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
)
from xgboost import XGBClassifier
from imblearn.over_sampling import SMOTE

# ════════════════════════════════════════════════════════════════════
# FEATURE SETS
# ════════════════════════════════════════════════════════════════════

df=pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Cardiovascular Disease /CVD_Cleandata May 2026 _File.xlsx")

TARGET = 'AMI'

df_Lasso = df.drop(columns=["HTN", "DM", "SMK", "FH", "Age_Group", "TG1", "TC1"])
df_Chi   = df.drop(columns=["TC1", "DL", "TG1", "FH"])
df_Boruta = df.drop(columns=['TG1','TC1','SMK','DL'])


datasets = {
    "F1 - Lasso Selected Feature"     : df_Lasso,
    "F2 - Chi-Square Selected Feature": df_Chi,
    "F3 - Boruta Selected Feature"    : df_Boruta,
}

NON_ENSEMBLE = ["CART", "SVM", "Naive Bayes", "KNN"]
ENSEMBLE     = ["Random Forest", "GBM", "AdaBoost", "XGBoost"]

# ════════════════════════════════════════════════════════════════════
# MODELS — exact hyper-parameters from manuscript Table 2
# ════════════════════════════════════════════════════════════════════
def get_models():
    return {
        "CART": DecisionTreeClassifier(
            criterion='gini', max_depth=2,
            min_samples_leaf=1, random_state=42),
        "SVM": SVC(
            kernel='rbf', C=10, gamma=0.1,
            probability=True, random_state=42),
        "Naive Bayes": BernoulliNB(alpha=10.0),
        "KNN": KNeighborsClassifier(
            n_neighbors=4, weights='distance',
            algorithm='auto', p=1),
        "Random Forest": RandomForestClassifier(
            n_estimators=400, max_depth=10,
            min_samples_leaf=2, random_state=42),
        "GBM": GradientBoostingClassifier(
            learning_rate=0.1, max_depth=3,
            n_estimators=100, random_state=42),
        "AdaBoost": AdaBoostClassifier(
            learning_rate=0.0001, n_estimators=10,
            random_state=42),
        "XGBoost": XGBClassifier(
            learning_rate=0.02, max_depth=5,
            min_child_weight=0.05, n_estimators=100,
            random_state=42, eval_metric='logloss', verbosity=0),
    }

# ════════════════════════════════════════════════════════════════════
# HELPER — McNemar significance label
# ════════════════════════════════════════════════════════════════════
def significance_label(p):
    if   p < 0.001: return "p<0.001 Highly Significant"
    elif p < 0.01:  return "p<0.01 Very Significant"
    elif p < 0.05:  return "p<0.05 Significant"
    else:           return "Not Significant"

# ════════════════════════════════════════════════════════════════════
# HELPER — Distribution bar plot
# ════════════════════════════════════════════════════════════════════
def plot_distribution(axes_list, labels_list, titles, colors_list, suptitle):
    fig, axes = plt.subplots(1, len(axes_list), figsize=(10, 3))
    fig.suptitle(suptitle, fontsize=13, fontweight='bold')
    for ax, labels, title, colors in zip(axes, labels_list, titles, colors_list):
        unique, counts = np.unique(labels, return_counts=True)
        bars = ax.bar([str(u) for u in unique], counts,
                      color=colors[:len(unique)], edgecolor='black')
        ax.set_title(title)
        ax.set_xlabel("Class")
        ax.set_ylabel("Count")
        for bar, cnt in zip(bars, counts):
            ax.text(bar.get_x() + bar.get_width() / 2,
                    bar.get_height(), f'{cnt:,}',
                    ha='center', va='bottom', fontsize=10)
    plt.tight_layout()
    plt.show()

# ════════════════════════════════════════════════════════════════════
# MAIN PIPELINE
# ════════════════════════════════════════════════════════════════════
def run_pipeline(data_df, feature_set_name, n_splits=10, random_state=42):

    X = data_df.drop(columns=TARGET).values
    y = data_df[TARGET].values.ravel()
    model_names = list(get_models().keys())

    print("\n" + "═" * 65)
    print(f"  {feature_set_name}")
    print(f"  Features : {X.shape[1]}  |  Samples : {len(y):,}")
    print("═" * 65)
    print(f"\n  X shape : {X.shape}")
    print(f"  Y shape : {y.shape}")
    print(f"\n  Class distribution:")
    for cls, cnt in zip(*np.unique(y, return_counts=True)):
        print(f"    Class {cls} : {cnt:,}  ({cnt/len(y)*100:.1f}%)")

    # ════════════════════════════════════════════════════════════════
    # VISUALISATION STEP 1 — Train / Test Split Plot
    # (illustrative 80/20 split for plotting only)
    # ════════════════════════════════════════════════════════════════
    print("\n  ── Train / Test Split (illustrative 80/20 for plot) ──")
    X_vis_tr, X_vis_te, y_vis_tr, y_vis_te = train_test_split(
        X, y, test_size=0.20, random_state=random_state, stratify=y
    )
    print(f"  Train : {len(y_vis_tr):,} | Test : {len(y_vis_te):,}")
    for cls, cnt in zip(*np.unique(y_vis_tr, return_counts=True)):
        print(f"    Train Class {cls} : {cnt:,}  ({cnt/len(y_vis_tr)*100:.1f}%)")
    for cls, cnt in zip(*np.unique(y_vis_te, return_counts=True)):
        print(f"    Test  Class {cls} : {cnt:,}  ({cnt/len(y_vis_te)*100:.1f}%)")

    plot_distribution(
        axes_list   = [0, 1],
        labels_list = [y_vis_tr,              y_vis_te],
        titles      = ["Train Set",           "Test Set (Sealed)"],
        colors_list = [["#818cf8","#f472b6"], ["#fbbf24","#f87171"]],
        suptitle    = f"{feature_set_name} — Train / Test Distribution"
    )

    # ════════════════════════════════════════════════════════════════
    # VISUALISATION STEP 2 — Before vs After SMOTE Plot
    # ════════════════════════════════════════════════════════════════
    print("\n  ── Before vs After SMOTE (on illustrative train set) ──")
    scaler_vis      = StandardScaler()
    X_vis_scaled    = scaler_vis.fit_transform(X_vis_tr)
    sm_vis          = SMOTE(random_state=random_state)
    _, y_vis_res    = sm_vis.fit_resample(X_vis_scaled, y_vis_tr)

    print(f"  Before SMOTE : {len(y_vis_tr):,} → After SMOTE : {len(y_vis_res):,}")
    for cls, cnt in zip(*np.unique(y_vis_res, return_counts=True)):
        print(f"    Class {cls} : {cnt:,}  ({cnt/len(y_vis_res)*100:.1f}%)")

    plot_distribution(
        axes_list   = [0, 1],
        labels_list = [y_vis_tr,                y_vis_res],
        titles      = ["Before SMOTE (Train)",  "After SMOTE (Train)"],
        colors_list = [["#818cf8","#f472b6"],   ["#818cf8","#34d399"]],
        suptitle    = f"{feature_set_name} — Before vs After SMOTE"
    )

    # ════════════════════════════════════════════════════════════════
    # STRATIFIED K-FOLD CV
    # SMOTE applied strictly inside each training fold only
    # ════════════════════════════════════════════════════════════════
    skf = StratifiedKFold(n_splits=n_splits, shuffle=True,
                          random_state=random_state)

    fold_scores = {name: {m: [] for m in
                   ['accuracy', 'precision', 'recall', 'f1', 'auc', 'auprc']}
                   for name in model_names}
    oof_preds   = {name: np.zeros(len(y), dtype=int) for name in model_names}
    oof_true    = np.zeros(len(y), dtype=int)

    print(f"\n  Running {n_splits}-fold Stratified CV "
          f"(SMOTE inside each train fold only)...\n")

    for fold, (train_idx, test_idx) in enumerate(skf.split(X, y), 1):

        X_train, X_test = X[train_idx], X[test_idx]
        y_train, y_test = y[train_idx], y[test_idx]

        # Scale — fit on train only, transform both
        scaler  = StandardScaler()
        X_train = scaler.fit_transform(X_train)
        X_test  = scaler.transform(X_test)          # ← no leakage

        # SMOTE — train fold only, test fold never touched
        sm           = SMOTE(random_state=random_state)
        X_res, y_res = sm.fit_resample(X_train, y_train)

        oof_true[test_idx] = y_test                 # store true labels

        for name, model in get_models().items():
            model.fit(X_res, y_res)
            y_pred = model.predict(X_test)
            y_prob = model.predict_proba(X_test)[:, 1]

            oof_preds[name][test_idx] = y_pred      # store for McNemar

            fold_scores[name]['accuracy'].append(
                accuracy_score(y_test, y_pred))
            fold_scores[name]['precision'].append(
                precision_score(y_test, y_pred, average='macro', zero_division=0))
            fold_scores[name]['recall'].append(
                recall_score(y_test, y_pred, average='macro', zero_division=0))
            fold_scores[name]['f1'].append(
                f1_score(y_test, y_pred, average='macro', zero_division=0))
            fold_scores[name]['auc'].append(
                roc_auc_score(y_test, y_prob))
            fold_scores[name]['auprc'].append(
                average_precision_score(y_test, y_prob))

        print(f"  Fold {fold:02d}/{n_splits} ✓")

    # ════════════════════════════════════════════════════════════════
    # TABLE 1 — Mean ± SD
    # ════════════════════════════════════════════════════════════════
    sd_rows = []
    for name in model_names:
        s   = fold_scores[name]
        cat = "Non-Ensemble" if name in NON_ENSEMBLE else "Ensemble"
        sd_rows.append({
            'Category'             : cat,
            'Model'                : name,
            'Accuracy (Mean ± SD)' : f"{np.mean(s['accuracy']):.4f} ± {np.std(s['accuracy'],  ddof=1):.4f}",
            'Precision (Mean ± SD)': f"{np.mean(s['precision']):.4f} ± {np.std(s['precision'], ddof=1):.4f}",
            'Recall (Mean ± SD)'   : f"{np.mean(s['recall']):.4f} ± {np.std(s['recall'],    ddof=1):.4f}",
            'F1 Score (Mean ± SD)' : f"{np.mean(s['f1']):.4f} ± {np.std(s['f1'],       ddof=1):.4f}",
            'AUC Score (Mean ± SD)': f"{np.mean(s['auc']):.4f} ± {np.std(s['auc'],      ddof=1):.4f}",
        })
    sd_df = pd.DataFrame(sd_rows)

    print(f"\n{'═'*75}")
    print(f"  TABLE 1 — Mean ± SD  |  {feature_set_name}")
    print(f"{'═'*75}")
    print(sd_df.to_string(index=False))

    # ════════════════════════════════════════════════════════════════
    # TABLE 2 — Point Estimates + AUPRC
    # ════════════════════════════════════════════════════════════════
    acc_rows = []
    for name in model_names:
        s   = fold_scores[name]
        cat = "Non-Ensemble" if name in NON_ENSEMBLE else "Ensemble"
        acc_rows.append({
            'Category' : cat,
            'Model'    : name,
            'Accuracy' : round(np.mean(s['accuracy']),  2),
            'Precision': round(np.mean(s['precision']), 2),
            'Recall'   : round(np.mean(s['recall']),    2),
            'F1 Score' : round(np.mean(s['f1']),        2),
            'AUC Score': round(np.mean(s['auc']),       2),
            'AUPRC'    : round(np.mean(s['auprc']),     2),
        })
    acc_df = pd.DataFrame(acc_rows)

    print(f"\n{'═'*75}")
    print(f"  TABLE 2 — Accuracy Table  |  {feature_set_name}")
    print(f"{'═'*75}")
    print(acc_df.to_string(index=False))

    # ════════════════════════════════════════════════════════════════
    # TABLE 3 — Full Pairwise McNemar's Test (all 28 pairs)
    # ════════════════════════════════════════════════════════════════
    mcn_rows = []
    for name_a, name_b in combinations(model_names, 2):
        pred_a = oof_preds[name_a]
        pred_b = oof_preds[name_b]
        b      = int(np.sum((pred_a == oof_true) & (pred_b != oof_true)))
        c      = int(np.sum((pred_a != oof_true) & (pred_b == oof_true)))
        result = mcnemar([[0, b], [c, 0]], exact=True)
        p      = result.pvalue
        mcn_rows.append({
            'Model A'           : name_a,
            'Model B'           : name_b,
            'b'                 : b,
            'c'                 : c,
            'p-value'           : round(p, 4),
            'Significance Level': significance_label(p),
        })
    mcn_df = pd.DataFrame(mcn_rows)

    print(f"\n{'═'*75}")
    print(f"  TABLE 3 — McNemar's Test  |  {feature_set_name}")
    print(f"{'═'*75}")
    print(mcn_df.to_string(index=False))

    return sd_df, acc_df, mcn_df


# ════════════════════════════════════════════════════════════════════
# RUN ALL 3 FEATURE SETS
# ════════════════════════════════════════════════════════════════════
all_results = {}

for fs_name, data_df in datasets.items():
    sd, acc, mcn         = run_pipeline(data_df, fs_name)
    all_results[fs_name] = {'sd': sd, 'acc': acc, 'mcn': mcn}


# ════════════════════════════════════════════════════════════════════
# EXPORT TO EXCEL — one sheet per table per feature set
# ════════════════════════════════════════════════════════════════════
with pd.ExcelWriter("CVD_Results.xlsx", engine="openpyxl") as writer:
    for fs_name, tbls in all_results.items():
        short = fs_name.split()[0]                        # Lasso / Chi / Boruta
        tbls['sd'].to_excel(writer,
            sheet_name=f"{short}_MeanSD",   index=False)
        tbls['acc'].to_excel(writer,
            sheet_name=f"{short}_Accuracy", index=False)
        tbls['mcn'].to_excel(writer,
            sheet_name=f"{short}_McNemar",  index=False)

print("\n  ✅ All results exported → CVD_Results.xlsx")
print("     Sheets: Lasso_MeanSD, Lasso_Accuracy, Lasso_McNemar,")
print("             Chi_MeanSD,   Chi_Accuracy,   Chi_McNemar,")
print("             Boruta_MeanSD, Boruta_Accuracy, Boruta_McNemar")
```

```python
# ════════════════════════════════════════════════════════════════════
# CVD DATASET — ROC · PRC · CONFUSION MATRIX
# Adapted from reference code for your cardiovascular pipeline
# ════════════════════════════════════════════════════════════════════

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import StratifiedKFold, train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    roc_curve, auc, precision_recall_curve,
    average_precision_score, confusion_matrix,
    classification_report
)
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import BernoulliNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import (
    RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
)
from xgboost import XGBClassifier
from imblearn.over_sampling import SMOTE
import warnings
warnings.filterwarnings('ignore')

# ════════════════════════════════════════════════════════════════════
# FEATURE SETS — your cardiovascular dataset
# ════════════════════════════════════════════════════════════════════
TARGET = 'AMI'

df_Lasso  = df.drop(columns=["TC1", "DL", "TG1", "FH"])
df_Chi    = df.drop(columns=["TC1", "DL", "TG1", "FH", "LDL1", "HDL1"])
df_Boruta = df[['HTN', 'DM', 'Sex', 'NSTEMI', 'SA', 'UA',
                'HOCABG', 'HOPCI', 'Age_Group', 'EF1',
                'Indication1', 'FH', 'LDL1', 'HDL1', 'AMI']]

datasets = {
    "F1 - Lasso Selected Feature"     : df_Lasso,
    "F2 - Chi-Square Selected Feature": df_Chi,
    "F3 - Boruta Selected Feature"    : df_Boruta,
}

# ════════════════════════════════════════════════════════════════════
# MODEL DISPLAY CONFIG
# ════════════════════════════════════════════════════════════════════
NON_ENSEMBLE = {
    "CART"       : "Classification and Regression Trees",
    "SVM"        : "Support Vector Machine",
    "Naive Bayes": "Naive Bayes",
    "KNN"        : "K-Nearest Neighbors",
}
ENSEMBLE = {
    "Random Forest": "Random Forest",
    "GBM"          : "Gradient Boosting Machine",
    "AdaBoost"     : "AdaBoost",
    "XGBoost"      : "XGBoost",
}

NON_ENS_COLORS = ['#3b82f6', '#ef4444', '#f59e0b', '#10b981']
ENS_COLORS     = ['#8b5cf6', '#06b6d4', '#f97316', '#ec4899']
LINESTYLES     = ['--', '-.', ':', '-']

# ════════════════════════════════════════════════════════════════════
# MODELS — exact hyper-parameters from manuscript Table 2
# ════════════════════════════════════════════════════════════════════
def get_models():
    return {
        "CART": DecisionTreeClassifier(
            criterion='gini', max_depth=2,
            min_samples_leaf=1, random_state=42),
        "SVM": SVC(
            kernel='rbf', C=10, gamma=0.1,
            probability=True, random_state=42),
        "Naive Bayes": BernoulliNB(alpha=10.0),
        "KNN": KNeighborsClassifier(
            n_neighbors=4, weights='distance',
            algorithm='auto', p=1),
        "Random Forest": RandomForestClassifier(
            n_estimators=400, max_depth=10,
            min_samples_leaf=2, random_state=42),
        "GBM": GradientBoostingClassifier(
            learning_rate=0.1, max_depth=3,
            n_estimators=100, random_state=42),
        "AdaBoost": AdaBoostClassifier(
            learning_rate=0.0001, n_estimators=10,
            random_state=42),
        "XGBoost": XGBClassifier(
            learning_rate=0.02, max_depth=5,
            min_child_weight=0.05, n_estimators=100,
            random_state=42, eval_metric='logloss', verbosity=0),
    }

# ════════════════════════════════════════════════════════════════════
# PIPELINE — SMOTE on train, collect OOF probs via 10-fold CV
# Using OOF probabilities gives smoother, more honest curves
# than a single train/test split
# ════════════════════════════════════════════════════════════════════
def run_pipeline_for_plots(data_df, feature_set_name,
                            n_splits=10, random_state=42):

    X = data_df.drop(columns=TARGET).values
    y = data_df[TARGET].values.ravel()
    model_names = list(get_models().keys())

    print(f"\n{'═'*65}")
    print(f"  {feature_set_name}  |  {X.shape[1]} features  |  n={len(y):,}")
    print(f"{'═'*65}")

    skf = StratifiedKFold(n_splits=n_splits, shuffle=True,
                          random_state=random_state)

    # Out-of-fold storage
    oof_probs = {name: np.zeros(len(y))      for name in model_names}
    oof_preds = {name: np.zeros(len(y), dtype=int) for name in model_names}
    oof_true  = np.zeros(len(y), dtype=int)

    for fold, (train_idx, test_idx) in enumerate(skf.split(X, y), 1):
        X_train, X_test = X[train_idx], X[test_idx]
        y_train, y_test = y[train_idx], y[test_idx]

        # Scale — fit on train only
        scaler  = StandardScaler()
        X_train = scaler.fit_transform(X_train)
        X_test  = scaler.transform(X_test)

        # SMOTE — train fold only
        sm           = SMOTE(random_state=random_state)
        X_res, y_res = sm.fit_resample(X_train, y_train)

        oof_true[test_idx] = y_test

        for name, model in get_models().items():
            model.fit(X_res, y_res)
            oof_preds[name][test_idx] = model.predict(X_test)
            oof_probs[name][test_idx] = model.predict_proba(X_test)[:, 1]

        print(f"  Fold {fold:02d}/{n_splits} ✓")

    # Compute AUC and AP from OOF predictions
    auc_scores = {}
    ap_scores  = {}
    for name in model_names:
        fpr, tpr, _ = roc_curve(oof_true, oof_probs[name])
        auc_scores[name] = auc(fpr, tpr)
        ap_scores[name]  = average_precision_score(oof_true, oof_probs[name])
        print(f"  {name:<18} | AUC: {auc_scores[name]:.4f} "
              f"| AP: {ap_scores[name]:.4f}")

    return oof_probs, oof_preds, oof_true, auc_scores, ap_scores


# ════════════════════════════════════════════════════════════════════
# FIGURE 1 — ROC CURVES  (Non-Ensemble | Ensemble side by side)
# ════════════════════════════════════════════════════════════════════
def plot_roc(oof_probs, oof_true, auc_scores, fs_name):

    fig, axes = plt.subplots(1, 2, figsize=(18, 7))
    fig.suptitle(f'ROC Curves — {fs_name}',
                 fontsize=15, fontweight='bold', y=1.02)

    for ax, model_map, colors, title in zip(
        axes,
        [NON_ENSEMBLE,          ENSEMBLE],
        [NON_ENS_COLORS,        ENS_COLORS],
        ["Non-Ensemble Models", "Ensemble Models"]
    ):
        best_key = max(model_map.keys(),
                       key=lambda k: auc_scores.get(k, 0))

        for (key, display), color, ls in zip(
            model_map.items(), colors, LINESTYLES
        ):
            fpr, tpr, _ = roc_curve(oof_true, oof_probs[key])
            auc_val     = auc_scores[key]
            ax.plot(fpr, tpr,
                    color=color, linestyle=ls, linewidth=2.5,
                    label=f'{display} (AUC = {auc_val:.3f})')

        # Shade under best model curve
        fpr_b, tpr_b, _ = roc_curve(oof_true, oof_probs[best_key])
        ax.fill_between(fpr_b, tpr_b, alpha=0.07, color='steelblue')

        # Random baseline
        ax.plot([0, 1], [0, 1], color='gray', linestyle='--',
                linewidth=1.2, label='Random Classifier (AUC = 0.500)')

        ax.set_title(title, fontsize=13, fontweight='bold', pad=10)
        ax.set_xlabel('False Positive Rate', fontsize=12)
        ax.set_ylabel('True Positive Rate',  fontsize=12)
        ax.legend(loc='lower right', fontsize=9, framealpha=0.9)
        ax.set_xlim([0.0, 1.0])
        ax.set_ylim([0.0, 1.05])
        ax.grid(True, linestyle='--', alpha=0.35)
        ax.set_facecolor('#f8fafc')
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)

    plt.tight_layout()
    fname = f"ROC_{fs_name.replace(' ','_')}.png"
    plt.savefig(fname, dpi=150, bbox_inches='tight')
    plt.show()
    print(f"  ✅ ROC saved → {fname}")


# ════════════════════════════════════════════════════════════════════
# FIGURE 2 — PRECISION-RECALL (AUPRC) CURVES
# ════════════════════════════════════════════════════════════════════
def plot_prc(oof_probs, oof_true, ap_scores, fs_name):

    baseline = np.sum(oof_true) / len(oof_true)

    fig, axes = plt.subplots(1, 2, figsize=(18, 7))
    fig.suptitle(f'Precision-Recall Curves — {fs_name}',
                 fontsize=15, fontweight='bold', y=1.02)

    for ax, model_map, colors, title in zip(
        axes,
        [NON_ENSEMBLE,          ENSEMBLE],
        [NON_ENS_COLORS,        ENS_COLORS],
        ["Non-Ensemble Models", "Ensemble Models"]
    ):
        best_key = max(model_map.keys(),
                       key=lambda k: ap_scores.get(k, 0))

        for (key, display), color, ls in zip(
            model_map.items(), colors, LINESTYLES
        ):
            prec, rec, _ = precision_recall_curve(oof_true, oof_probs[key])
            ap_val       = ap_scores[key]
            ax.plot(rec, prec,
                    color=color, linestyle=ls, linewidth=2.5,
                    label=f'{display} (AUPRC = {ap_val:.3f})')

        # Shade under best model curve
        prec_b, rec_b, _ = precision_recall_curve(
            oof_true, oof_probs[best_key])
        ax.fill_between(rec_b, prec_b, alpha=0.07, color='steelblue')

        # Random baseline
        ax.axhline(y=baseline, color='gray', linestyle='--',
                   linewidth=1.2,
                   label=f'Random Classifier (AUPRC = {baseline:.3f})')

        ax.set_title(title, fontsize=13, fontweight='bold', pad=10)
        ax.set_xlabel('Recall',    fontsize=12)
        ax.set_ylabel('Precision', fontsize=12)
        ax.legend(loc='upper right', fontsize=9, framealpha=0.9)
        ax.set_xlim([0.0, 1.0])
        ax.set_ylim([0.0, 1.05])
        ax.grid(True, linestyle='--', alpha=0.35)
        ax.set_facecolor('#f8fafc')
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)

    plt.tight_layout()
    fname = f"PRC_{fs_name.replace(' ','_')}.png"
    plt.savefig(fname, dpi=150, bbox_inches='tight')
    plt.show()
    print(f"  ✅ PRC saved → {fname}")


# ════════════════════════════════════════════════════════════════════
# FIGURE 3 — CONFUSION MATRICES
# 3a: Best model single large CM
# 3b: All 8 models in a 2×4 grid
# ════════════════════════════════════════════════════════════════════
def plot_confusion_matrices(oof_preds, oof_true, auc_scores, fs_name):

    best_model = max(auc_scores, key=auc_scores.get)
    model_names = list(oof_preds.keys())

    # ── 3a: Single best model CM ─────────────────────────────────
    fig, ax = plt.subplots(figsize=(6, 5))
    cm      = confusion_matrix(oof_true, oof_preds[best_model])
    cm_pct  = cm.astype(float) / cm.sum(axis=1, keepdims=True) * 100

    sns.heatmap(cm, annot=False, cmap='Blues', ax=ax,
                linewidths=0.5, linecolor='white',
                cbar_kws={'shrink': 0.8})

    for i in range(2):
        for j in range(2):
            text_color = 'white' if cm[i, j] > cm.max() * 0.5 else 'black'
            ax.text(j + 0.5, i + 0.38,
                    f'{cm[i, j]:,}',
                    ha='center', va='center',
                    fontsize=16, fontweight='bold',
                    color=text_color)
            ax.text(j + 0.5, i + 0.65,
                    f'({cm_pct[i, j]:.1f}%)',
                    ha='center', va='center',
                    fontsize=11, color=text_color)

    ax.set_xticklabels(['CVD Negative\n(Predicted)',
                         'CVD Positive\n(Predicted)'], fontsize=10)
    ax.set_yticklabels(['CVD Negative\n(Actual)',
                         'CVD Positive\n(Actual)'],
                        fontsize=10, rotation=0)
    ax.set_title(
        f'Confusion Matrix — {best_model} ⭐\n'
        f'{fs_name}  |  AUC = {auc_scores[best_model]:.4f}',
        fontsize=12, fontweight='bold', pad=12
    )

    # Metrics below the plot
    tn, fp, fn, tp = cm.ravel()
    sensitivity = tp / (tp + fn)
    specificity = tn / (tn + fp)
    ppv         = tp / (tp + fp)
    npv         = tn / (tn + fn)
    fig.text(0.5, -0.03,
             f"Sensitivity={sensitivity:.3f}   Specificity={specificity:.3f}"
             f"   PPV={ppv:.3f}   NPV={npv:.3f}",
             ha='center', fontsize=9, color='#475569')

    plt.tight_layout()
    fname = f"CM_best_{fs_name.replace(' ','_')}.png"
    plt.savefig(fname, dpi=150, bbox_inches='tight')
    plt.show()
    print(f"  ✅ Best model CM saved → {fname}")

    # ── 3b: All 8 models in 2×4 grid ─────────────────────────────
    fig, axes = plt.subplots(2, 4, figsize=(22, 11))
    fig.suptitle(
        f'Confusion Matrices — All Models  |  {fs_name}',
        fontsize=14, fontweight='bold'
    )

    for ax, name in zip(axes.flatten(), model_names):
        cm_i     = confusion_matrix(oof_true, oof_preds[name])
        cm_pct_i = cm_i.astype(float) / cm_i.sum(axis=1, keepdims=True) * 100

        sns.heatmap(cm_i, annot=False, cmap='Blues', ax=ax,
                    linewidths=0.5, linecolor='white', cbar=False)

        for i in range(2):
            for j in range(2):
                tc = 'white' if cm_i[i, j] > cm_i.max() * 0.5 else 'black'
                ax.text(j + 0.5, i + 0.38,
                        f'{cm_i[i, j]:,}',
                        ha='center', va='center',
                        fontsize=13, fontweight='bold', color=tc)
                ax.text(j + 0.5, i + 0.65,
                        f'({cm_pct_i[i, j]:.1f}%)',
                        ha='center', va='center',
                        fontsize=9, color=tc)

        star = " ⭐" if name == best_model else ""
        ax.set_title(
            f'{name}{star}\nAUC = {auc_scores[name]:.3f}',
            fontsize=10, fontweight='bold'
        )
        ax.set_xticklabels(['CVD–\n(Pred)', 'CVD+\n(Pred)'], fontsize=8)
        ax.set_yticklabels(['CVD–\n(Act)',  'CVD+\n(Act)'],
                            fontsize=8, rotation=0)
        ax.set_xlabel('Predicted', fontsize=8)
        ax.set_ylabel('Actual',    fontsize=8)

    plt.tight_layout()
    fname = f"CM_all_{fs_name.replace(' ','_')}.png"
    plt.savefig(fname, dpi=150, bbox_inches='tight')
    plt.show()
    print(f"  ✅ All models CM saved → {fname}")

    # Classification report for best model
    print(f"\n  Classification Report — {best_model} ({fs_name}):")
    print(f"  {'─'*55}")
    print(classification_report(
        oof_true, oof_preds[best_model],
        target_names=['CVD Negative (0)', 'CVD Positive (1)']
    ))

    return best_model


# ════════════════════════════════════════════════════════════════════
# RUN — ALL 3 FEATURE SETS
# ════════════════════════════════════════════════════════════════════
summary = []

for fs_name, data_df in datasets.items():

    print("\n" + "═" * 65)
    print(f"  Processing — {fs_name}")
    print("═" * 65)

    # Step 1 — Run CV pipeline, collect OOF probs & preds
    oof_probs, oof_preds, oof_true, auc_sc, ap_sc = \
        run_pipeline_for_plots(data_df, fs_name)

    # Step 2 — ROC Curves
    print(f"\n  Plotting ROC Curves...")
    plot_roc(oof_probs, oof_true, auc_sc, fs_name)

    # Step 3 — PRC / AUPRC Curves
    print(f"\n  Plotting PRC Curves...")
    plot_prc(oof_probs, oof_true, ap_sc, fs_name)

    # Step 4 — Confusion Matrices
    print(f"\n  Plotting Confusion Matrices...")
    best = plot_confusion_matrices(oof_preds, oof_true, auc_sc, fs_name)

    summary.append({
        'Feature Set': fs_name,
        'Best Model' : best,
        'AUC'        : round(auc_sc[best], 4),
        'AUPRC'      : round(ap_sc[best],  4),
    })

# ════════════════════════════════════════════════════════════════════
# FINAL SUMMARY
# ════════════════════════════════════════════════════════════════════
print("\n" + "═" * 65)
print("  BEST MODEL SUMMARY — All Feature Sets")
print("═" * 65)
print(pd.DataFrame(summary).to_string(index=False))
print("═" * 65)

print("""
  FILES SAVED:
  ROC_F1_-_Lasso_Selected_Feature.png
  ROC_F2_-_Chi-Square_Selected_Feature.png
  ROC_F3_-_Boruta_Selected_Feature.png

  PRC_F1_-_Lasso_Selected_Feature.png
  PRC_F2_-_Chi-Square_Selected_Feature.png
  PRC_F3_-_Boruta_Selected_Feature.png

  CM_best_F1_-_Lasso_Selected_Feature.png
  CM_best_F2_-_Chi-Square_Selected_Feature.png
  CM_best_F3_-_Boruta_Selected_Feature.png

  CM_all_F1_-_Lasso_Selected_Feature.png
  CM_all_F2_-_Chi-Square_Selected_Feature.png
  CM_all_F3_-_Boruta_Selected_Feature.png
""")
```

# Without SMOTE

```python
# ════════════════════════════════════════════════════════════════════
# CARDIOVASCULAR DISEASE PREDICTION — WITH vs WITHOUT SMOTE PIPELINE
# Produces parallel results for both conditions to show SMOTE impact
# ════════════════════════════════════════════════════════════════════

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from itertools import combinations
from statsmodels.stats.contingency_tables import mcnemar
from sklearn.model_selection import StratifiedKFold, train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, roc_auc_score, average_precision_score
)
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import BernoulliNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import (
    RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
)
from xgboost import XGBClassifier
from imblearn.over_sampling import SMOTE

# ════════════════════════════════════════════════════════════════════
# FEATURE SETS
# ════════════════════════════════════════════════════════════════════

df = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Cardiovascular Disease /CVD_Cleandata May 2026 _File.xlsx")

TARGET = 'AMI'

df_Lasso  = df.drop(columns=["HTN", "DM", "SMK", "FH", "Age_Group", "TG1", "TC1"])
df_Chi    = df.drop(columns=["TC1", "DL", "TG1", "FH"])
df_Boruta = df.drop(columns=['TG1', 'TC1', 'SMK', 'DL'])

datasets = {
    "F1 - Lasso Selected Feature"     : df_Lasso,
    "F2 - Chi-Square Selected Feature": df_Chi,
    "F3 - Boruta Selected Feature"    : df_Boruta,
}

NON_ENSEMBLE = ["CART", "SVM", "Naive Bayes", "KNN"]
ENSEMBLE     = ["Random Forest", "GBM", "AdaBoost", "XGBoost"]

# ════════════════════════════════════════════════════════════════════
# MODELS
# ════════════════════════════════════════════════════════════════════
def get_models():
    return {
        "CART": DecisionTreeClassifier(
            criterion='gini', max_depth=2,
            min_samples_leaf=1, random_state=42),
        "SVM": SVC(
            kernel='rbf', C=10, gamma=0.1,
            probability=True, random_state=42),
        "Naive Bayes": BernoulliNB(alpha=10.0),
        "KNN": KNeighborsClassifier(
            n_neighbors=4, weights='distance',
            algorithm='auto', p=1),
        "Random Forest": RandomForestClassifier(
            n_estimators=400, max_depth=10,
            min_samples_leaf=2, random_state=42),
        "GBM": GradientBoostingClassifier(
            learning_rate=0.1, max_depth=3,
            n_estimators=100, random_state=42),
        "AdaBoost": AdaBoostClassifier(
            learning_rate=0.0001, n_estimators=10,
            random_state=42),
        "XGBoost": XGBClassifier(
            learning_rate=0.02, max_depth=5,
            min_child_weight=0.05, n_estimators=100,
            random_state=42, eval_metric='logloss', verbosity=0),
    }

# ════════════════════════════════════════════════════════════════════
# HELPERS
# ════════════════════════════════════════════════════════════════════
def significance_label(p):
    if   p < 0.001: return "p<0.001 Highly Significant"
    elif p < 0.01:  return "p<0.01 Very Significant"
    elif p < 0.05:  return "p<0.05 Significant"
    else:           return "Not Significant"


def plot_distribution(labels_list, titles, colors_list, suptitle):
    fig, axes = plt.subplots(1, len(labels_list), figsize=(10, 3))
    fig.suptitle(suptitle, fontsize=13, fontweight='bold')
    for ax, labels, title, colors in zip(axes, labels_list, titles, colors_list):
        unique, counts = np.unique(labels, return_counts=True)
        bars = ax.bar([str(u) for u in unique], counts,
                      color=colors[:len(unique)], edgecolor='black')
        ax.set_title(title)
        ax.set_xlabel("Class")
        ax.set_ylabel("Count")
        for bar, cnt in zip(bars, counts):
            ax.text(bar.get_x() + bar.get_width() / 2,
                    bar.get_height(), f'{cnt:,}',
                    ha='center', va='bottom', fontsize=10)
    plt.tight_layout()
    plt.show()


# ════════════════════════════════════════════════════════════════════
# CORE CV LOOP — use_smote flag controls SMOTE on/off
# ════════════════════════════════════════════════════════════════════
def run_cv(X, y, n_splits=10, random_state=42, use_smote=True):
    """
    Returns fold_scores dict and oof_preds / oof_true arrays.
    use_smote=True  → SMOTE applied inside every training fold
    use_smote=False → raw (imbalanced) training fold used directly
    """
    model_names = list(get_models().keys())
    skf = StratifiedKFold(n_splits=n_splits, shuffle=True,
                          random_state=random_state)

    fold_scores = {name: {m: [] for m in
                   ['accuracy', 'precision', 'recall', 'f1', 'auc', 'auprc']}
                   for name in model_names}
    oof_preds = {name: np.zeros(len(y), dtype=int) for name in model_names}
    oof_true  = np.zeros(len(y), dtype=int)

    label = "WITH SMOTE" if use_smote else "WITHOUT SMOTE"
    print(f"\n  Running {n_splits}-fold Stratified CV [{label}]...\n")

    for fold, (train_idx, test_idx) in enumerate(skf.split(X, y), 1):

        X_train, X_test = X[train_idx], X[test_idx]
        y_train, y_test = y[train_idx], y[test_idx]

        # Scale — fit on train only
        scaler  = StandardScaler()
        X_train = scaler.fit_transform(X_train)
        X_test  = scaler.transform(X_test)

        # Conditionally apply SMOTE
        if use_smote:
            sm           = SMOTE(random_state=random_state)
            X_res, y_res = sm.fit_resample(X_train, y_train)
        else:
            X_res, y_res = X_train, y_train   # no resampling

        oof_true[test_idx] = y_test

        for name, model in get_models().items():
            model.fit(X_res, y_res)
            y_pred = model.predict(X_test)
            y_prob = model.predict_proba(X_test)[:, 1]

            oof_preds[name][test_idx] = y_pred

            fold_scores[name]['accuracy'].append(
                accuracy_score(y_test, y_pred))
            fold_scores[name]['precision'].append(
                precision_score(y_test, y_pred, average='macro', zero_division=0))
            fold_scores[name]['recall'].append(
                recall_score(y_test, y_pred, average='macro', zero_division=0))
            fold_scores[name]['f1'].append(
                f1_score(y_test, y_pred, average='macro', zero_division=0))
            fold_scores[name]['auc'].append(
                roc_auc_score(y_test, y_prob))
            fold_scores[name]['auprc'].append(
                average_precision_score(y_test, y_prob))

        print(f"    Fold {fold:02d}/{n_splits} ✓")

    return fold_scores, oof_preds, oof_true


# ════════════════════════════════════════════════════════════════════
# BUILD RESULT TABLES FROM fold_scores
# ════════════════════════════════════════════════════════════════════
def build_tables(fold_scores):
    model_names = list(fold_scores.keys())
    sd_rows, acc_rows = [], []

    for name in model_names:
        s   = fold_scores[name]
        cat = "Non-Ensemble" if name in NON_ENSEMBLE else "Ensemble"

        sd_rows.append({
            'Category'             : cat,
            'Model'                : name,
            'Accuracy (Mean ± SD)' : f"{np.mean(s['accuracy']):.4f} ± {np.std(s['accuracy'],  ddof=1):.4f}",
            'Precision (Mean ± SD)': f"{np.mean(s['precision']):.4f} ± {np.std(s['precision'], ddof=1):.4f}",
            'Recall (Mean ± SD)'   : f"{np.mean(s['recall']):.4f} ± {np.std(s['recall'],    ddof=1):.4f}",
            'F1 Score (Mean ± SD)' : f"{np.mean(s['f1']):.4f} ± {np.std(s['f1'],       ddof=1):.4f}",
            'AUC Score (Mean ± SD)': f"{np.mean(s['auc']):.4f} ± {np.std(s['auc'],      ddof=1):.4f}",
        })
        acc_rows.append({
            'Category' : cat,
            'Model'    : name,
            'Accuracy' : round(np.mean(s['accuracy']),  4),
            'Precision': round(np.mean(s['precision']), 4),
            'Recall'   : round(np.mean(s['recall']),    4),
            'F1 Score' : round(np.mean(s['f1']),        4),
            'AUC Score': round(np.mean(s['auc']),       4),
            'AUPRC'    : round(np.mean(s['auprc']),     4),
        })

    return pd.DataFrame(sd_rows), pd.DataFrame(acc_rows)


# ════════════════════════════════════════════════════════════════════
# BUILD McNemar TABLE
# ════════════════════════════════════════════════════════════════════
def build_mcnemar(oof_preds, oof_true):
    model_names = list(oof_preds.keys())
    mcn_rows = []
    for name_a, name_b in combinations(model_names, 2):
        pred_a = oof_preds[name_a]
        pred_b = oof_preds[name_b]
        b      = int(np.sum((pred_a == oof_true) & (pred_b != oof_true)))
        c      = int(np.sum((pred_a != oof_true) & (pred_b == oof_true)))
        result = mcnemar([[0, b], [c, 0]], exact=True)
        p      = result.pvalue
        mcn_rows.append({
            'Model A'           : name_a,
            'Model B'           : name_b,
            'b'                 : b,
            'c'                 : c,
            'p-value'           : round(p, 4),
            'Significance Level': significance_label(p),
        })
    return pd.DataFrame(mcn_rows)


# ════════════════════════════════════════════════════════════════════
# IMPACT SUMMARY — delta between With vs Without SMOTE
# ════════════════════════════════════════════════════════════════════
def build_impact_summary(acc_with, acc_without):
    """
    Computes per-model delta (With SMOTE − Without SMOTE) for each metric.
    Positive delta means SMOTE improved that metric.
    """
    metrics = ['Accuracy', 'Precision', 'Recall', 'F1 Score', 'AUC Score', 'AUPRC']
    rows = []
    for _, row_w in acc_with.iterrows():
        row_wo = acc_without[acc_without['Model'] == row_w['Model']].iloc[0]
        entry  = {'Category': row_w['Category'], 'Model': row_w['Model']}
        for m in metrics:
            delta = round(row_w[m] - row_wo[m], 4)
            entry[f'Δ {m}'] = f"{delta:+.4f}"
        rows.append(entry)
    return pd.DataFrame(rows)


# ════════════════════════════════════════════════════════════════════
# FULL PIPELINE — runs both conditions for one feature set
# ════════════════════════════════════════════════════════════════════
def run_pipeline(data_df, feature_set_name, n_splits=10, random_state=42):

    X = data_df.drop(columns=TARGET).values
    y = data_df[TARGET].values.ravel()

    print("\n" + "═" * 65)
    print(f"  {feature_set_name}")
    print(f"  Features : {X.shape[1]}  |  Samples : {len(y):,}")
    print("═" * 65)
    for cls, cnt in zip(*np.unique(y, return_counts=True)):
        print(f"  Class {cls} : {cnt:,}  ({cnt/len(y)*100:.1f}%)")

    # ── Visualisation: Train / Test split ──────────────────────────
    X_vis_tr, X_vis_te, y_vis_tr, y_vis_te = train_test_split(
        X, y, test_size=0.20, random_state=random_state, stratify=y)
    plot_distribution(
        labels_list = [y_vis_tr,              y_vis_te],
        titles      = ["Train Set",           "Test Set (Sealed)"],
        colors_list = [["#818cf8","#f472b6"], ["#fbbf24","#f87171"]],
        suptitle    = f"{feature_set_name} — Train / Test Distribution"
    )

    # ── Visualisation: Before vs After SMOTE ───────────────────────
    scaler_vis   = StandardScaler()
    X_vis_scaled = scaler_vis.fit_transform(X_vis_tr)
    sm_vis       = SMOTE(random_state=random_state)
    _, y_vis_res = sm_vis.fit_resample(X_vis_scaled, y_vis_tr)
    plot_distribution(
        labels_list = [y_vis_tr,               y_vis_res],
        titles      = ["Before SMOTE (Train)", "After SMOTE (Train)"],
        colors_list = [["#818cf8","#f472b6"],  ["#818cf8","#34d399"]],
        suptitle    = f"{feature_set_name} — Before vs After SMOTE"
    )

    # ── CV WITHOUT SMOTE ───────────────────────────────────────────
    fs_wo, preds_wo, true_wo = run_cv(X, y, n_splits, random_state,
                                      use_smote=False)
    sd_wo, acc_wo = build_tables(fs_wo)
    mcn_wo        = build_mcnemar(preds_wo, true_wo)

    print(f"\n{'═'*75}")
    print(f"  WITHOUT SMOTE — Mean ± SD  |  {feature_set_name}")
    print(f"{'═'*75}")
    print(sd_wo.to_string(index=False))

    print(f"\n{'═'*75}")
    print(f"  WITHOUT SMOTE — Accuracy Table  |  {feature_set_name}")
    print(f"{'═'*75}")
    print(acc_wo.to_string(index=False))

    print(f"\n{'═'*75}")
    print(f"  WITHOUT SMOTE — McNemar's Test  |  {feature_set_name}")
    print(f"{'═'*75}")
    print(mcn_wo.to_string(index=False))

    # ── CV WITH SMOTE ──────────────────────────────────────────────
    fs_w, preds_w, true_w = run_cv(X, y, n_splits, random_state,
                                   use_smote=True)
    sd_w, acc_w = build_tables(fs_w)
    mcn_w       = build_mcnemar(preds_w, true_w)

    print(f"\n{'═'*75}")
    print(f"  WITH SMOTE — Mean ± SD  |  {feature_set_name}")
    print(f"{'═'*75}")
    print(sd_w.to_string(index=False))

    print(f"\n{'═'*75}")
    print(f"  WITH SMOTE — Accuracy Table  |  {feature_set_name}")
    print(f"{'═'*75}")
    print(acc_w.to_string(index=False))

    print(f"\n{'═'*75}")
    print(f"  WITH SMOTE — McNemar's Test  |  {feature_set_name}")
    print(f"{'═'*75}")
    print(mcn_w.to_string(index=False))

    # ── SMOTE Impact Summary ───────────────────────────────────────
    impact = build_impact_summary(acc_w, acc_wo)

    print(f"\n{'═'*75}")
    print(f"  SMOTE IMPACT (With − Without)  |  {feature_set_name}")
    print(f"  (+) = SMOTE improved  |  (−) = SMOTE hurt  |  0 = no change")
    print(f"{'═'*75}")
    print(impact.to_string(index=False))

    return {
        'sd_without'  : sd_wo,  'acc_without' : acc_wo,  'mcn_without': mcn_wo,
        'sd_with'     : sd_w,   'acc_with'    : acc_w,   'mcn_with'   : mcn_w,
        'smote_impact': impact,
    }


# ════════════════════════════════════════════════════════════════════
# RUN ALL 3 FEATURE SETS
# ════════════════════════════════════════════════════════════════════
all_results = {}

for fs_name, data_df in datasets.items():
    all_results[fs_name] = run_pipeline(data_df, fs_name)


# ════════════════════════════════════════════════════════════════════
# EXPORT TO EXCEL
# Sheet naming:
#   <FS>_NoSMOTE_MeanSD / _Accuracy / _McNemar
#   <FS>_SMOTE_MeanSD   / _Accuracy / _McNemar
#   <FS>_SMOTE_Impact
# ════════════════════════════════════════════════════════════════════
with pd.ExcelWriter("CVD_Results_SMOTE_Comparison.xlsx",
                    engine="openpyxl") as writer:

    short_map = {
        "F1 - Lasso Selected Feature"     : "Lasso",
        "F2 - Chi-Square Selected Feature": "Chi",
        "F3 - Boruta Selected Feature"    : "Boruta",
    }

    for fs_name, tbls in all_results.items():
        short = short_map[fs_name]

        tbls['sd_without'].to_excel(
            writer, sheet_name=f"{short}_NoSMOTE_MeanSD",   index=False)
        tbls['acc_without'].to_excel(
            writer, sheet_name=f"{short}_NoSMOTE_Accuracy",  index=False)
        tbls['mcn_without'].to_excel(
            writer, sheet_name=f"{short}_NoSMOTE_McNemar",   index=False)

        tbls['sd_with'].to_excel(
            writer, sheet_name=f"{short}_SMOTE_MeanSD",      index=False)
        tbls['acc_with'].to_excel(
            writer, sheet_name=f"{short}_SMOTE_Accuracy",    index=False)
        tbls['mcn_with'].to_excel(
            writer, sheet_name=f"{short}_SMOTE_McNemar",     index=False)

        tbls['smote_impact'].to_excel(
            writer, sheet_name=f"{short}_SMOTE_Impact",      index=False)

print("\n  ✅ All results exported → CVD_Results_SMOTE_Comparison.xlsx")
print("""
     Sheets created (7 per feature set × 3 sets = 21 sheets):
       Lasso_NoSMOTE_MeanSD    Chi_NoSMOTE_MeanSD    Boruta_NoSMOTE_MeanSD
       Lasso_NoSMOTE_Accuracy  Chi_NoSMOTE_Accuracy  Boruta_NoSMOTE_Accuracy
       Lasso_NoSMOTE_McNemar   Chi_NoSMOTE_McNemar   Boruta_NoSMOTE_McNemar
       Lasso_SMOTE_MeanSD      Chi_SMOTE_MeanSD      Boruta_SMOTE_MeanSD
       Lasso_SMOTE_Accuracy    Chi_SMOTE_Accuracy    Boruta_SMOTE_Accuracy
       Lasso_SMOTE_McNemar     Chi_SMOTE_McNemar     Boruta_SMOTE_McNemar
       Lasso_SMOTE_Impact      Chi_SMOTE_Impact      Boruta_SMOTE_Impact
""")
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from imblearn.over_sampling import SMOTE

# ─────────────────────────────────────────────
# 1. Load your dataset
# ─────────────────────────────────────────────
df=pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Cardiovascular Disease /CVD_Cleandata May 2026 _File.xlsx")

# Boruta-confirmed features (14 confirmed from Table 4)
boruta_features = [
    'HTN', 'Indication1', 'EF1', 'Age_Group', 'HOPCI',
    'UA', 'SA', 'HOCABG', 'Sex', 'DM', 'NSTEMI', 'FH', 'LDL1', 'HDL1'
]

target = 'AMI'

# --- FIX: Ensure feature columns are numeric and handle missing values ---
# This step should come after loading the data and defining features,
# but before splitting or using the data for model training.
for feature in boruta_features:
    # Convert column to numeric, coercing any non-numeric values to NaN
    df[feature] = pd.to_numeric(df[feature], errors='coerce')
    # Fill NaN values with the mean of the column.
    # For categorical-like features (e.g., 'Sex', 'HTN', 'DM' if they are binary/ordinal),
    # a different imputation strategy might be preferred (e.g., mode or 0/1 depending on context).
    # For now, mean is a general approach.
    if df[feature].isnull().all(): # If the entire column became NaN (e.g., all were non-numeric)
        df[feature] = df[feature].fillna(0) # FIX: Avoid inplace=True with chained assignment
    else:
        df[feature] = df[feature].fillna(df[feature].mean()) # FIX: Avoid inplace=True with chained assignment
# -------------------------------------------------------------------------


# ─────────────────────────────────────────────
# 2. Best model (AdaBoost with tuned params)
# ─────────────────────────────────────────────
model = AdaBoostClassifier(learning_rate=0.0001, n_estimators=10, random_state=42)

# ─────────────────────────────────────────────
# 3. Helper: evaluate one subgroup with SMOTE + stratified 10-fold CV
# ─────────────────────────────────────────────
def evaluate_subgroup(subset, features, target, label):
    X = subset[features].values
    y = subset[target].values

    skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
    metrics = {'accuracy': [], 'precision': [], 'recall': [], 'f1': [], 'auc': []}

    for fold, (train_idx, test_idx) in enumerate(skf.split(X, y), 1):
        X_train, X_test = X[train_idx], X[test_idx]
        y_train, y_test = y[train_idx], y[test_idx]

        # Apply SMOTE only on training fold (correct — no leakage)
        # Skip SMOTE if minority class has fewer than 2 samples in fold
        unique, counts = np.unique(y_train, return_counts=True)
        if len(unique) < 2 or min(counts) < 2:
            print(f"  [{label}] Fold {fold}: skipping SMOTE (too few minority samples)")
            X_res, y_res = X_train, y_train
        else:
            smote = SMOTE(random_state=42)
            X_res, y_res = smote.fit_resample(X_train, y_train)

        model.fit(X_res, y_res)
        y_pred = model.predict(X_test)
        y_prob = model.predict_proba(X_test)[:, 1]

        metrics['accuracy'].append(accuracy_score(y_test, y_pred) * 100)
        metrics['precision'].append(precision_score(y_test, y_pred, zero_division=0) * 100)
        metrics['recall'].append(recall_score(y_test, y_pred, zero_division=0) * 100)
        metrics['f1'].append(f1_score(y_test, y_pred, zero_division=0) * 100)
        try:
            metrics['auc'].append(roc_auc_score(y_test, y_prob))
        except ValueError:
            metrics['auc'].append(np.nan)

    print(f"\n{'='*55}")
    print(f"  Sex-Stratified Sensitivity Analysis — {label}")
    print(f"  n = {len(subset)} | CVD+ = {y.sum()} | CVD- = {(y==0).sum()}")
    print(f"{'='*55}")
    print(f"  {'Metric':<12} {'Mean':>8} {'Std':>8} {'95% CI':>20}")
    print(f"  {'-'*50}")
    for m, vals in metrics.items():
        arr = np.array(vals)
        mean = np.nanmean(arr)
        std  = np.nanstd(arr)
        ci_lo = mean - 1.96 * std / np.sqrt(len(arr))
        ci_hi = mean + 1.96 * std / np.sqrt(len(arr))
        unit = "%" if m != "auc" else ""
        print(f"  {m.capitalize():<12} {mean:>7.2f}{unit}  {std:>6.2f}  [{ci_lo:.2f}, {ci_hi:.2f}]")

    return {m: np.nanmean(v) for m, v in metrics.items()}

# ─────────────────────────────────────────────
# 4. Split by sex and run analysis
# ─────────────────────────────────────────────
# In your dataset: Sex = 1 → Female, Sex = 0 → Male (based on inspection)
female_df = df[df['Sex'] == 1].copy()
male_df   = df[df['Sex'] == 0].copy() # Corrected: Sex=0 for male based on data inspection

print(f"\nDataset split:")
print(f"  Female (Sex=1): n={len(female_df)}")
print(f"  Male   (Sex=0): n={len(male_df)}") # Updated print statement

female_results = evaluate_subgroup(female_df, boruta_features, target, "Female (Sex=1)")
male_results   = evaluate_subgroup(male_df,   boruta_features, target, "Male   (Sex=0)") # Updated label

# ─────────────────────────────────────────────
# 5. Side-by-side comparison table
# ─────────────────────────────────────────────
print(f"\n{'='*55}")
print("  Comparison: Full Dataset vs. Sex Subgroups")
print(f"{'='*55}")
print(f"  {'Metric':<12} {'Female':>10} {'Male':>10}")
print(f"  {'-'*35}")
for m in ['accuracy', 'precision', 'recall', 'f1', 'auc']:
    f_val = female_results[m]
    m_val = male_results[m]
    unit = "%" if m != "auc" else ""
    print(f"  {m.capitalize():<12} {f_val:>9.2f}{unit}  {m_val:>9.2f}{unit}")
```
